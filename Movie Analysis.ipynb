```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Import the data
mov = pd.read_csv('Movie_Analysis.csv', encoding = 'latin1')
mov.head()
```

```python
# Exploring the dataset
len(mov)
mov.columns = ['Day', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'Budget', 'Gross',
       'IMDbRating', 'MovieLensRating', 'Overseas', 'Overseas%',
       'Profit', 'Profit%', 'Runtime', 'US',
       'Gross%']
mov.columns
```

```python
# Checking the summary of the dataframe
mov.info()
```

```python
# Checking the structure of the dataframe
mov.describe()
```

```python
# Exploreing the categorical variable Studio, used in the assignment
mov.Studio = mov.Studio.astype("category")
```

```python
# Exploring the categorical variable Studio, used in the assignment
mov.Genre._()
```

```python
# Filteing the dataframe by genre
mov2 = mov[(mov.Genre == 'action') | (mov.Genre == 'adventure') | (mov.Genre == 'animation') | (mov.Genre == 'comedy') | (mov.Genre == 'drama')]
```

```python
# Filtering the dataframe by studio
mov3 = mov2[(mov2.Studio == 'Buena Vista Studios') | (mov2.Studio == 'Fox') | (mov2.Studio == 'Paramount Pictures') | (mov2. Studio == 'Sony') | (mov2.Studio == 'Universal') | (mov2.Studio == 'WB')]
```

```python
# Checking how the filters worked
print (mov.Genre.unique())
print (mov.Studio.unique())
print (len(mov))
```

```python
# Defineing the style
sns.set(style="darkgrid", palette="muted", color_codes=True)
plt.rcParams["figure.figsize"] = 10,10

# Ploting the boxsplots
ax = sns.boxplot(data=mov, x='Genre', y='Gross%', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Adding in points to show each observation
sns.stripplot(x='Genre', y='Gross%', data=mov, jitter=True, size=6, linewidth=0, hue = 'Studio', alpha=0.7)

ax.axes.set_title('Domestic Gross by Genre',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross%',fontsize=20)

# Defineing where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.
         )
```
